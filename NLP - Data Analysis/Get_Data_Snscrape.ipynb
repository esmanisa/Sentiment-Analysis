## Get Data

```python
query = 'trendyol Trendyol (#trendyol OR #Trendyol) lang:tr until:2023-03-18 since:2006-01-01'
tweets = []

get_tweets = sntwitter.TwitterSearchScraper(query).get_items()

limit = 7000

for tweet in get_tweets:
  if len(tweets) == limit:
    break
  else:
    tweets.append([tweet.date, tweet.user.username, tweet.content])

df_trendyol = pd.DataFrame(tweets, columns=['Date', 'UserName', 'tweets'])

df_trendyol['Date'] = df_trendyol['Date'].apply(lambda a: pd.to_datetime(a).date())

print(df_trendyol.duplicated().sum())

df_trendyol.drop_duplicates(inplace=True)

df_trendyol.to_csv("twitter_data_trendyol.csv")
df_trendyol.to_excel("twitter_data_trendyol.xlsx")
```
